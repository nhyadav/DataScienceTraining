```python
#data types

print(type("123"))
print(type('abcd')) #string always consider into both the coutes ' or "

print(type(123)) #int

print(type(True)) #bolean

print(type(3.4)) #float

print(type(3+5j)) #complex

print(type([1,2,3])) # sequence data type---list
print(type((1,2,3))) # sequence datatype --- tuple
print(type({'a':1,'b':3})) #sequence datatype---dict
print(type({1,2,3,4})) #sequence datatype --- set
print(type(None))
```

## Keywords in the Python
